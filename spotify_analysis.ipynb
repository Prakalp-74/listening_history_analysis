```python
df["track_name"] = df["track_name"].fillna("Unknown")
df["artist"] = df["artist"].fillna("Unknown")
df["album"] = df["album"].fillna("Unknown")
df["track_uri"] = df["track_uri"].fillna("Unknown")
df["reason_start"] = df["reason_start"].fillna("Unknown") 
df["reason_end"] = df["reason_end"].fillna("Unknown")                                                           
```

```python
skipped_tracks = (df["skipped"].sum()/len(df)) * 100
```

```python
print("Precentage of skipped tracks", skipped_tracks)
```

```python
percentage_shuffle = ((df["shuffle"].sum()/len(df)) * 100 ).round(2)
```

```python
print("Precentage of shuffling", percentage_shuffle)
```
